```python
import numpy as np
```

```python
#x: 描画したい理論値の定義域
#y: 実験したいものの理論値
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
#横軸縦軸の範囲を決めれます
plt.xlim(0,2*np.pi)
plt.ylim(-2,2)
#縦軸横軸の名前を出せます
plt.xlabel("x")
plt.ylabel("sin(x)")
#gridが欲しかったら出してください
plt.grid()
#Titleも出せます, y引数で場所を調整できます
plt.title('xとsin(x)の関係', y = -0.2)
plt.plot(x, y, label = "理論値")
#凡例を出すなら下を忘れずに
plt.legend()
plt.show()
```

```python
#例として実験値モドキも出してみる
y_exp = np.sin(x) + (np.random.rand(len(x)) - 0.5)/5*3
#(np.random.rand(len(x)) - 0.5)とすることで[-0.5, 0.5]の間での乱数が出てくれる。それを3/5倍しているので[-0.3,0.3の間で値が出ます。]
```

```python
plt.xlim(0,2*np.pi)
plt.ylim(-2,2)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid()
plt.title('xとsin(x)の関係', y = -0.2)
plt.plot(x, y,label = "理論値")
#試しにscatterで描画、引数のsは記号の大きさ、colorは色、markerは記号の形を指定している
plt.scatter(x, y_exp,label = "実験値",s=20,color = "red",marker = "x")
plt.legend()
plt.show()
```

```python
#少しエラーバー入りのグラフを書きたいとき、例えば測定に0.1の誤差が生じるらしいことがわかっている時を考えます
y_err = np.zeros(len(x)) + 0.1
```

```python
plt.xlim(0,2*np.pi)
plt.ylim(-2,2)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid()
plt.title('xとsin(x)の関係', y = -0.2)
plt.plot(x, y,label = "理論値")
plt.errorbar(x, y_exp,yerr = y_err,label = "実験値",markersize=4,marker = "x",color = "red",capsize = 5,linestyle="None")
plt.legend()
plt.show()
```
