```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib #matplotlib単体では日本語を出力できない
from matplotlib import font_manager
#システム内のフォントを探してくる下はルートディレクトリのSystem/Library/Fonts/Library/Fonts内部にあるフォントファイルを探してきている
font_manager.fontManager.addfont("/System/Library/Fonts/ヒラギノ明朝 ProN.ttc")
```

```python
fontnames = [f.name for f in font_manager.fontManager.ttflist]
```

```python
#以下のコードを実行すればmatplotlibで使えるフォントの一覧が出力できる
import codecs
with codecs.open('fonts.txt', 'w', 'utf-8') as f:
    for i in range(len(fontnames)):
        f.write(fontnames[i]+"\n")
    f.close()
#生成したファイル(fonts.txtができているはず)を見る限り、ヒラギノ明朝はHiragino Mincho ProNとして登録されているみたい
```

```python
#x: 描画したい理論値の定義域
#y: 実験したいものの理論値
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
#実際に
plt.rcParams['font.family'] = 'Hiragino Mincho ProN'
#横軸縦軸の範囲を決めれます
plt.xlim(0,2*np.pi)
plt.ylim(-2,2)
#縦軸横軸の名前を出せます、fontnameが引数に含まれているmatplotlib関数は個別にフォントを指定できます
#また$$で囲むことでlatex表記が使えます。
plt.xlabel(r"$x$",math_fontfamily='cm')
plt.ylabel(r"$\sin x$",math_fontfamily='cm')
#gridが欲しかったら出してください
#gridもcolorとalphaを指定できます
plt.grid(color = "gray",alpha = 0.2)
#Titleも出せます, y引数で場所を調整できます
plt.title('$x$と$\sin x$の関係', y = -0.2,fontname = "Hiragino Mincho ProN",math_fontfamily='cm')
#linewidthでプロットの時の線の太さも決められます
plt.plot(x, y,color = "black",linewidth = 1)
#凡例を出すなら下を忘れずに、フォントはpropに辞書型の配列を渡します
plt.legend(["理論値$Graph$"], prop={"family":"Hiragino Mincho ProN", "math_fontfamily":"cm"})
plt.show()
```
